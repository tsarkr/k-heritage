```python
import pandas as pd
import matplotlib.pyplot as plt

# 파일 경로
file_path = 'youtube_comments.csv'

# 데이터 읽기
data = pd.read_csv(file_path)

# 댓글 작성 시간 열 확인 및 변환
if 'published_at' in data.columns:
    data['published_at'] = pd.to_datetime(data['published_at'])  # datetime 형식으로 변환
else:
    raise ValueError("댓글 작성 시간 열 ('published_at')이 데이터에 없습니다.")

# 연도별 댓글 수 집계
data['year'] = data['published_at'].dt.year  # 연도 추출
yearly_counts = data['year'].value_counts().sort_index()  # 연도별 댓글 수 정렬

# 히스토그램 생성
plt.figure(figsize=(10, 6))
plt.bar(yearly_counts.index, yearly_counts.values, color='skyblue', edgecolor='black')
plt.xlabel('Year', fontsize=12)
plt.ylabel('Number of Comments', fontsize=12)
plt.title('Yearly Distribution of Comments', fontsize=14)
plt.xticks(yearly_counts.index, rotation=45)  # 연도 표시
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.tight_layout()
plt.show()
```

```python
from wordcloud import WordCloud
import matplotlib.pyplot as plt
import pandas as pd

# 파일 경로
file_path = 'extracted_nouns_konlpy.csv'

# 데이터 읽기
data = pd.read_csv(file_path)

# 데이터 준비 (Keyword와 Frequency를 딕셔너리로 변환)
word_freq = dict(zip(data['Keyword'], data['Frequency']))

# 워드 클라우드 생성
font_path = "/System/Library/Fonts/Supplemental/AppleGothic.ttf"  # macOS에서 한글 폰트 경로
wordcloud = WordCloud(
    font_path=font_path,
    width=800,
    height=400,
    background_color='white',
    colormap='viridis',
    max_words=200
).generate_from_frequencies(word_freq)

# 워드 클라우드 시각화
plt.figure(figsize=(12, 6))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis("off")
plt.title("WordCloud for Extracted Nouns", fontsize=16)
plt.show()
```
